```python
import torch
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
with open("./model_weights/pre_trained/unet/postpretrained-unet-values", "rb") as f:
    saved_lists = pickle.load(f)
    
epoch_loss_values, val_epoch_loss_values, train_metric_values, metric_values, best_metric, best_metric_epoch = saved_lists
```

```python
val_interval = 2

plt.figure("train", (12, 6))
plt.subplot(1, 2, 1)
plt.title("Epoch Average Loss")
x = [i + 1 for i in range(len(epoch_loss_values))]
y = epoch_loss_values
z = val_epoch_loss_values
plt.xlabel("epoch")
plt.ylim(0.25, 0.65)
plt.plot(x, y, label = "training")
plt.plot(x, z, label = "validation")
plt.legend()
plt.subplot(1, 2, 2)
plt.title("Val Mean Dice")
x = [val_interval * (i + 1) for i in range(len(metric_values))]
y = metric_values
z = train_metric_values
plt.xlabel("epoch")
plt.plot(x, z, label = "training")
plt.plot(x, y, label = "validation")
plt.legend()
# plt.savefig('pretrained-unet-figures.png')
plt.show()
```

```python
print(
    f"train completed, best_metric: {best_metric:.4f} "
    f"at epoch: {best_metric_epoch}")
```
